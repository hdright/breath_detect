```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
noBpmPoints = 600
BPMresol = 0.1
cfg = {'Np':3, 'gt':np.array([10, 13, 30])}
cfg = {'Np':1, 'gt':np.array([15])}
```

```python
gt_pd = np.zeros(noBpmPoints)
sigma2 = (1 / BPMresol ** 2)
x = np.arange(noBpmPoints)
for p in range(cfg['Np']):
    # gt所有值的高斯分布叠加作为label
    mu = cfg['gt'][p] / BPMresol
    gt_pd += np.exp(-(x - mu) ** 2 / (2 * sigma2)) / np.sqrt(2 * np.pi * sigma2) / cfg['Np']
```

```python
plt.figure()
plt.plot(gt_pd)
plt.show()
```
